#### QUESTION 2: Read and parse data from the```"ride.csv"```. Complete the given code (8 marks)

````python
import csv

def load_ride_data(filename):
    data = []
    with open(filename, mode='r') as file:
        reader = csv.DictReader(file)
        # read and parse data iteratively  (5 marks)
        for row in reader:
            data.append(row)
    return data

# Write code(s) to load data from actual CSV file (2 marks)
ride_data = load_ride_data('ride.csv')

# Write code(s) to preview data (1 mark)
for row in ride_data:
    print(row)
````

#### QUESTION 3: Define a function to compute average fare by city and call the function by passing the data (10 marks)

````python
# Write your function definition here to compute average fare by city (8 marks)
def compute_average_fare_by_city(data):
    city_fares = {}
    city_counts = {}

    for row in data:
        city = row['city']
        fare = float(row['fare'])  # convert fare from string to float

        if city in city_fares:
            city_fares[city] += fare
            city_counts[city] += 1
        else:
            city_fares[city] = fare
            city_counts[city] = 1

    average_fares = {}
    for city in city_fares:
        average_fares[city] = city_fares[city] / city_counts[city]

    return average_fares

# Write the code(s) to call your function and pass the data (2 marks)
avg_fares = compute_average_fare_by_city(ride_data)
print(avg_fares)
````

#### QUESTION 4: Define a function to summarize driver information, call the function by passing data ad display the summary (10 marks)
#### ```driver_name, num_of_rides, total_distance, total_fare```

````python
# Write your function definition here (7 marks)
def summarize_driver_info(data):
    driver_summary = {}

    for row in data:
        driver = row['driver_name']
        distance = float(row['distance'])
        fare = float(row['fare'])

        if driver in driver_summary:
            driver_summary[driver]['num_of_rides'] += 1
            driver_summary[driver]['total_distance'] += distance
            driver_summary[driver]['total_fare'] += fare
        else:
            driver_summary[driver] = {
                'num_of_rides': 1,
                'total_distance': distance,
                'total_fare': fare
            }

    return driver_summary

# Write the code(s) to call your function and pass the data (2 marks)
driver_info = summarize_driver_info(ride_data)

# Write the code(s) to display driver's summary (1 mark)
print("Driver Name | Number of Rides | Total Distance | Total Fare")
for driver, info in driver_info.items():
    print(f"{driver} | {info['num_of_rides']} | {info['total_distance']} | {info['total_fare']}")
````

#### QUESTION 5: Implement a function ```analyze_peak_slice(data, start_idx, end_idx)``` that: (10 marks)

    1.Accepts the ride data list and two indices: start_idx and end_idx.
    2.Extracts a slice of rides from start_idx = 2 to end_idx = 6 (inclusive).
    3.Calculates and returns:
        i)  Total number of rides in that slice
        ii) Average fare
        iii)List of unique drivers involved
        
    Sample output:
 
   ```Total_rides: 5```<br>
   ```Average_fare: 18.45```<br>
   ```Drivers: Ravi, Ali, Mei``` 

````python
# Write your function definition here that accepts ride data, start index and end index, slice the list, 
# calculate and returns the values in (3)   (8 marks)
def analyze_peak_slice(data, start_idx, end_idx):
    sliced_data = data[start_idx:end_idx + 1]  # include end_idx

    total_rides = len(sliced_data)

    total_fare = 0
    drivers_set = set()

    for row in sliced_data:
        total_fare += float(row['fare'])
        drivers_set.add(row['driver_name'])

    average_fare = total_fare / total_rides if total_rides > 0 else 0
    drivers_list = list(drivers_set)

    return total_rides, average_fare, drivers_list

# Write the code(s) to call your function and pass the ride data, start index and end index (2 marks)
total_rides, avg_fare, drivers = analyze_peak_slice(ride_data, 2, 6)

# Display the result (optional for clarity, but safe to include)
print(f"Total_rides: {total_rides}")
print(f"Average_fare: {round(avg_fare, 2)}")
print(f"Drivers: {', '.join(drivers)}")
````

#### QUESTION 6: Use lambda function to find top 3 drivers by total fare (7 marks)

````python
# Write your function definition here for top 3 drivers by total fare (5 marks)
def top_3_drivers_by_fare(driver_summary):
    # Convert dictionary into a list of tuples (driver_name, total_fare)
    driver_fares = [(driver, info['total_fare']) for driver, info in driver_summary.items()]

    # Sort list by total_fare in descending order using lambda
    sorted_drivers = sorted(driver_fares, key=lambda x: x[1], reverse=True)

    # Return top 3
    return sorted_drivers[:3]

# Write the code(s) to call your function and pass the driver's summary data (2 marks)
top_3 = top_3_drivers_by_fare(driver_info)

# Display the result (optional for clarity, but safe to include)
print("Top 3 drivers by total fare:")
for driver, fare in top_3:
    print(f"{driver}: {fare}")
````

#### QUESTION 7: Write a function to calculate average duration by cities and use lambda function to sort cities by average duration (10 marks)

````python
# Write your function definition here for average duration and sort cities (8 marks)
def average_duration_by_city(data):
    city_stats = {}  # to hold total duration and count

    for row in data:
        city = row['city']
        duration = float(row['duration'])

        if city not in city_stats:
            city_stats[city] = {'total_duration': 0, 'count': 0}
        
        city_stats[city]['total_duration'] += duration
        city_stats[city]['count'] += 1

    # Compute average duration per city
    city_avg_duration = []
    for city, stats in city_stats.items():
        avg_duration = stats['total_duration'] / stats['count']
        city_avg_duration.append( (city, avg_duration) )

    # Sort cities by average duration using lambda
    sorted_cities = sorted(city_avg_duration, key=lambda x: x[1], reverse=True)

    return sorted_cities

# Write the code(s) to call your function and pass the data (2 marks)
sorted_avg_duration = average_duration_by_city(ride_data)

# Display the result (optional for clarity)
print("Cities sorted by average duration:")
for city, avg_duration in sorted_avg_duration:
    print(f"{city}: {round(avg_duration, 2)} minutes")
````

#### QUESTION 8: Define a function to write driver summary to a text file ```driver_summary.txt``` with the following information:
#### ```"Driver", "Total Rides", "Total Distance", "Total Fare"```

````python
# Write your function definition here for writing driver summary into a text file (8 marks)
def write_driver_summary(filename, driver_summary):
    with open(filename, mode='w') as file:
        # Write header
        file.write("Driver\tTotal Rides\tTotal Distance\tTotal Fare\n")
        
        # Write each driver's info
        for driver, info in driver_summary.items():
            line = f"{driver}\t{info['num_of_rides']}\t{info['total_distance']}\t{info['total_fare']}\n"
            file.write(line)

# Write the code(s) to call your function and pass the driver summary data (2 marks)
write_driver_summary("driver_summary.txt", driver_info)
````
