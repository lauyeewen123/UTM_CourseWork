```python
#Question 1(a)

def calculate_BMI( weight, height):
    bmi = weight / (height* height)
    if bmi >= 30:
        return "Obesity"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Underweight"

print(calculate_BMI(50,1.6))
```

```python
import numpy as np

random_num = np.random.randint(1,101, size = 30)
random_num
```

```python
arr= random_num.reshape(5,6)
print(arr)
```

```python
col3 = arr[:, 2]
print("Column 3:", col3)
```

```python
col4 = arr[:, 3]
sum_col4 = np.sum(col4)
print("Sum of Column 4:", sum_col4)
```

```python
sum_col3_col4 = col3 + col4
print("Sum of Col3 and Col4 (element-wise):", sum_col3_col4)
```

```python
print("Matrix:\n", arr)
print()
print("Column 3:", col3)
print("Column 4:", col4)
print("Sum of Column 4:", sum_col4)
print("Element-wise sum of Col3 and Col4:", sum_col3_col4)
```
